```python

# Local do arquivo
Caminho = r'/content/PEDE_PASSOS_DATASET_FIAP_AJUSTADO VF.xlsx'
df = pd.read_excel(Caminho, sheet_name="VF")

# Lista de colunas que devem ser formatadas
colunas_tupla = ('IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN', 'INDE', 'PONTO_VIRADA')

# Garantir que apenas as colunas existentes no DataFrame sejam selecionadas
df = df[[col for col in colunas_tupla if col in df.columns]]

# Remover linhas onde 'PONTO_VIRADA' é 'Sem dados'
df = df[df['PONTO_VIRADA'] != 'Sem dados']

# Preencher NaN com 0
df = df.fillna(0)

# Converter colunas numéricas para float
colunas_numericas = ('IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN', 'INDE')

for col in colunas_numericas:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace(',', '.').astype(float)

# Exibir informações do DataFrame atualizado
df.info()

```

```python
# importa a base para o h2o
df_h2o=h2o.H2OFrame(df)
```

```python
#Transforma a variável dependente em categorica
df_h2o ["PONTO_VIRADA"]=df_h2o ["PONTO_VIRADA"].asfactor()
```

```python
#separa em treino e teste
treino, teste = df_h2o.split_frame(ratios=[0.7], seed=12)
```

```python
#Cria o modelo

ModeloAutoML=H2OAutoML(max_runtime_secs=180, sort_metric='AUC')
ModeloAutoML.train(y='PONTO_VIRADA', training_frame=treino)
```

```python
#Verifica as melhores curvas ROC
Ranking=ModeloAutoML.leaderboard
Ranking=Ranking.as_data_frame()
Ranking
```

```python
leaderboard = ModeloAutoML.leaderboard
leaderboard.head()
```

```python
# 1. Obtenha o ID do modelo treinado na leaderboard usando a base de desenvolvimento
model_id = ModeloAutoML.leaderboard[0, 'model_id']

# 2. Carregue o modelo treinado usando esse ID
modelo_carregado = h2o.get_model(model_id)

# 3. Avalie o desempenho do modelo na base de teste
performance_teste = modelo_carregado.model_performance(test_data=teste)
performance_teste
```
